# Removing Outliers

```python
# We will separate the features having outliers to apply zscore on it. 

outliers=df[['R','ERA','SHO','SV','E']]
```

```python
# We will treat these outliers using Zscore method.

from scipy.stats import zscore
z=np.abs(zscore(outliers))
df_new=df[(z<3).all(axis=1)]
df_new
```

We finally removed the outliers from the dataset.

# Performing Cross Validation for each algorithm.

```python
np.random.seed(10)
def rmse_cv(model,x,y):
    rmse =-(cross_val_score(model,x,y, scoring='neg_mean_squared_error', cv=10))
    return(rmse)

models = [LinearRegression(),
         SVR(kernel='linear'),
         RandomForestRegressor(),
         DecisionTreeRegressor(),
         KNN()]

names = ['lr','svr','rf','dt','knn']

for model,name in zip(models,names):
    score = rmse_cv(model,x,y)
    print("{}   : {:.6f}, {:4f}".format(name,score.mean(),score.std()))
```

According to cross validation's standard devivation score we got least std. score for randomforest regression algorithm i.e. 25.20 as compared to other algorithms. Lower the standard deviation score lower will be the variance thats why I am choosing Linear Regression as my final model.
